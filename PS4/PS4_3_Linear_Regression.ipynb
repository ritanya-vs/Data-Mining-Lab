Single Linear Regression

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("ps4_3_data.csv")
df.head()
```

```python
x_train = df.iloc[:,0].values
y_train = df.iloc[:,-1].values

x_mean = np.mean(x_train)
y_mean = np.mean(y_train)

#y = mx + b
numerator = np.sum((x_train - x_mean)*(y_train - y_mean))
denom = np.sum((x_train - x_mean)**2)
m = numerator/denom
b = y_mean - m*x_mean
```

```python
print("Single Linear Regression")
print("Slope (m):",m)
print("Intercept (b): ",b)

x_test = np.array([1,2])
y_pred = m*x_test + b
print("Test Set:",x_test)
print("Prediction:",y_pred)
```

Multiple Linear Regression

```python
x_train = df.iloc[:,:-1].values
y_train = df.iloc[:,-1].values
```

```python
ones = np.ones((x_train.shape[0],1))
x_b = np.hstack([ones,x_train])

XTX = np.dot(x_b.T,x_b)
XTX_inv = np.linalg.inv(XTX)
XTy = np.dot(x_b.T,y_train)
w = np.dot(XTX_inv,XTy)
print(w)
```

```python
print("Multiple Linear Regression:")
test_set = np.array([
    [2,3,4],
    [5,6,7]
])

ones = np.ones((test_set.shape[0],1))
x_test = np.hstack([ones,test_set])
y_pred = np.dot(x_test,w)
print("Test Set:",test_set)
print(y_pred)
```
